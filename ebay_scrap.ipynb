```python
## code to download the html file of a corresponding search term on ebay's website ###

search_term = "lg tv"

words = ("+").join((("").join(list(search_term))).split(" "))
url = "https://www.ebay.com/sch/i.html?_from=R40&_trksid=p2380057.m570.l1313&_nkw=" + words + "&_sacat=0"


r = requests.get(url ,headers=header)
soup = BeautifulSoup(r.text, 'lxml')

with open("ebay_lg_phone_01.htm", "w") as file:
    file.write(str(soup))    
```

```python
### looping through to download the html files of the first 10 pages ###


count = 1

while (count < 11):
    time.sleep(10)
    
    try:
    
        url_page = url + "&_pgn=" + str(count)
        r = requests.get(url_page ,headers=header)
        soup = BeautifulSoup(r.text, 'lxml')

        file_name = "ebay_lg_phone_"+str(count)+".htm"
        with open(file_name, "w") as file:
            file.write(str(soup))   
    
    except:
        break
    
    count += 1
```
